```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python


plt.title('first graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
axis_dict = {'fontname': 'Arial', 'fontsize': 15}
plt.xlabel('X axis', axis_dict)
plt.ylabel('Y axis', axis_dict)

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,7,7.5,8,10])
x = [0,1,2,3,4,5]
y = [0,2,4,6,8,10]
plt.plot(x,y, label='2x',c='red', linewidth=1.75, linestyle='--', marker='.', markersize=10, markeredgecolor='blue')
#plt.plot(x,y, 'r.--', label='2x')

x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:5], x2[:5]**2, label='x^2')
plt.plot(x2[4:], x2[4:]**2, linestyle='--')

plt.legend()
plt.figure(figsize=(5,3), dpi=300)

#plt.savefig('graph_one.png', dpi=300)


plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

#bars[0].set_hatch('/')
#bars[1].set_hatch('O')
#bars[2].set_hatch('*')
patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.figure(figsize=(6,4))

plt.show()
```

### Load Fiffa Data

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head()
```

#### Histograms

```python
fifa.shape[0]
```

```python
bins =[x for x in range(0,110,10)]
plt.hist(fifa.Overall, bins=bins, color='#abcdef')

plt.xticks(bins)

plt.ylabel('Numbers of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')

plt.show()
```

#### Pie Chart 1

```python
left = fifa[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.pie([left, right], labels = ['Left', 'Right'], autopct='%.2f %%')
plt.title('Foot Preference of FIFA Players')

plt.show()
```

#### Pie Chart 2

```python
fifa.Weight = [int(s.strip('lbs')) if type(s) == str else s for s in fifa.Weight]
```

```python
light = fifa[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight > 200].count()[0]
weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['< 125', '125 - 150', '150 - 175', '175 - 200', '> 200']
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,10))
plt.title('Weight Distribution of FIFA Players in lbs')
plt.pie(weights,labels=labels, autopct='%.2f %%', pctdistance=0.5, explode=(.4,.2,0,0,.4))
plt.show()
```

### Boxes

```python
labels = ['FC Barcelona', 'Real Madrid', 'New England Revolution']
barca = fifa[fifa.Club == labels[0]]['Overall']
madrid = fifa[fifa.Club == labels[1]]['Overall']
revs = fifa[fifa.Club == labels[2]]['Overall']
barca
```

```python
plt.style.use('default')
plt.figure(figsize=(8,8))

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Score')

boxes = plt.boxplot([barca, madrid, revs], labels=labels, patch_artist=True)
for box in boxes['boxes']:
    box.set(color='#4286f4', linewidth=2)
    box.set(facecolor='#e0e0e0')

plt.show()
```
